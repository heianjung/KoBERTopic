-  zero-shot Topic Modeling 
    - 일반적으로 AI가 뭔가를 탐지하려면 먼저 그것의 예시를 많이 보고 학습해야 해요. 예를 들어 "고양이 사진 찾기"를 하려면 수천 장의 고양이 사진을 보고 학습하죠. 하지만 **제로샷(Zero-shot)**은 **"예시를 전혀 안 봤는데도 찾을 수 있다"**는 뜻이에요.
    - 왜 유용한가요?
        - 새로운 이슈에 빠르게 대응: 갑자기 나타난 새로운 사회 이슈나 기술을 즉시 모니터링 가능
        - 전문가의 직관 활용: 해당 분야 전문가가 "이런 걸 주의 깊게 봐야 해"라고 설명하면, AI가 그걸 바로 찾아줌
        - 유연성: 관심 주제가 바뀔 때마다 다시 학습할 필요 없이, 설명만 바꾸면 됨
        - 쉽게 말해서 "이런 거 찾아줘"라고 말로만 설명하면, AI가 알아서 찾아주는 기술이에요!
